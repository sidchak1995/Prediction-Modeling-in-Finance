```python
def BlackScholes (option, K, T, S0, sigma, r, q):
    
    d1 = (np.log(S0/K)+(r-q+0.5*sigma**2)*T)/(sigma*T**0.5)
    d2 = d1 - sigma*T**0.5
    
    if option == 'C':
        price = (S0*np.e(-q*T)*norm.cdf(d1)) - K*np.e(-r*T)*norm.cdf(d2)
        
    if option == 'P':
        price = -(S0*np.e(-q*T)*norm.cdf(-d1)) + K*np.e(-r*T)*norm.cdf(-d2)
        
    return price
```
